```python
R = 1.0
c = 3e8
l = 0.01
```

```python
def random_walk_to_surface(R, l):
    position = np.array([0.0, 0.0])  # Start at the center (r = 0)
    path = []  # To store the path of the photon
    path.append(position.copy())
    while np.sqrt((position[0]**2) + (position[1])**2) < R:
      theta = np.random.uniform(0,2*np.pi)
      step = l * np.array([np.cos(theta),np.sin(theta)])
      position += step
      path.append(position.copy())
    return np.array(path)
```

```python
def plotrandomwalk(R,l):
  path = random_walk_to_surface(R,l)
  plt.plot(path[:,0],path[:,1])
  theta1 = np.linspace(0,2*np.pi,100)
  x = np.cos(theta1)
  y = np.sin(theta1)
  plt.plot(x,y)
  steps = len(path)
  plt.title(f'Random walk for l = 1 cm, R = 1m,  %d steps'% steps)
```

```python
plotrandomwalk(1,0.5)
```
